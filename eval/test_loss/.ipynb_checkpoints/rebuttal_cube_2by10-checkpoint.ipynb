```python
from utils.loss import ChamfersDistance3
import utils.utils as utils
import os
import numpy as np
import torch
import tqdm
import matplotlib.pyplot as plt
import json
```

```python
def load_ptcloud(folder_path, name, idx):
    """return ptcloud load from the path
    Params:
    -----------------------------------
    folder_path: string
        result folder path
    name: string
        name of ptcloud file 
    idx: int
        file idx
        
    Returns:
    -----------------------------------
    ptcloud: (ptnum,3) Numpy array
        point cloud loaded from this path
    """
    
    name = name[:name.find('.')] + '_{:03}'.format(idx) + '.npy'
    path = os.path.join(folder_path, name)
    ptcloud = np.load(path)
    return ptcloud
```

```python
def compute_ptcloud_dismatrix(X1, X2, distance_metric, title, results_dir, ifsave=False):
    """return distance matrix between ptcloud X1 ptcloud X2
    Params:
    ----------------------------------
    X1: (ptnum,3) torch.tensor
        point cloud 1
    X2: (ptnum,3) torch.tensor
        point cloud 2
    distance_metric: func
        metric to measure the distance of two point cloud
    ifsave: boolean
        if to save the distance matrix to disk
    title: string
        name of the saved matrix
    results_dir:string
        the path to save the distance matrix
        
    Returns:
    ----------------------------------
    D: (ptnum, ptnum) torch.tensor
        distance matrix
    """
    N = X1.shape[0]
    # initialize distance matrix
    D = np.zeros([N, N])
    # iterate over one group of ptcloud
    for i in range(N):	
        for j in range(N):
            D[i,j] = distance_metric(X1[i].unsqueeze(0), X2[j].unsqueeze(0))
    if ifsave:
        np.save(os.path.join(results_dir,title), D)
        print("saved to" + os.path.join(results_dir, title)) 
    return D
```

```python
def get_interp_list(data_dir = '../interp_data/cubesphere_1000/', test_json = 'test_interp_1000.json'):
    """return test set sample index and corresponding shapekey 
    Params:
    ------------------
    data_dir: string
        path to the data folder
    test_json: string
        path to the json file
    Returns: 
    ------------------
    interp_list: list of tuple 
        [(index, shapekey),...]
    """
    json_file = os.path.join(data_dir, test_json)
    data_path = open(json_file, 'r')
    testdata = json.load(data_path)
    interp_list = []
    for i in range(len(testdata)):
        index_key_pair = (i, testdata[i]['shapekey_value'])
        interp_list.append(index_key_pair)
        interp_list.sort(key=lambda pair: pair[1])
    return interp_list
```

```python
## Declare path and file name 
FOLDER = 'experiment/cubesphere_2by10_resnet/'
GT_NAME = 'oriptcloud.npy'
PRED_NAME = 'fineptcloud.npy'
device = torch.device("cuda")
chamfer = ChamfersDistance3()
chamfer = chamfer.to(device)
```

```python
## Pipeline to load the ptcloud
folder_idx_list = [1, 5, 10, 15, 20, 30, 35]  # seven experiments 
file_idx_list = [0, 1]                        # results stored as two npy file for one experiment
cube2by10_gt_list =[]                         # container
cube2by10_pred_list = []                      # container
dismatrix_list = []                           # container
interp_list = get_interp_list()               # Get the shapekey list 

## Loop through 7 experiments
for t, folder_idx in enumerate(folder_idx_list):
## Loop through all npy file in one experiment
    for i, file_idx in enumerate(file_idx_list):
        if i == 0:
            gt = load_ptcloud(os.path.join(FOLDER, 'results_{}/final_vis'.format(folder_idx)), GT_NAME, file_idx)
            pred = load_ptcloud(os.path.join(FOLDER, 'results_{}/final_vis'.format(folder_idx)), PRED_NAME, file_idx)
        else:
            gt = np.concatenate((gt, load_ptcloud(os.path.join(FOLDER, 'results_{}/final_vis'.format(folder_idx)), GT_NAME, file_idx)), axis =0)
            pred = np.concatenate((pred, load_ptcloud(os.path.join(FOLDER, 'results_{}/final_vis'.format(folder_idx)), PRED_NAME, file_idx)), axis =0)


    ## Sort the ptcloud tensor with the order of shape key
    pred = torch.from_numpy(pred)    # Numpy array to tensor (200, 1024, 3)
    gt = torch.from_numpy(gt)        # Numpy array to tensor (200, 1024, 3)
    sorted_gt = torch.zeros_like(gt)
    sorted_pred = torch.zeros_like(pred)
    for i in range(sorted_gt.shape[0]):
        sorted_gt[i] = gt[interp_list[i][0]]
        sorted_pred[i] = pred[interp_list[i][0]]
    cube2by10_gt_list.append(sorted_gt)
    cube2by10_pred_list.append(sorted_pred)
    
    ## Pipeline to compute the distance matrix between pred and gt
for idx in range(len(cube2by10_pred_list)):
    gt =  cube2by10_gt_list[idx]
    pred = cube2by10_pred_list[idx]
    gt = gt.to(device)
    pred = pred.to(device)
    dismatrix = compute_ptcloud_dismatrix(gt, pred, chamfer, title = '2by10_pred2gt_{}'.format(idx), 
                                          results_dir = 'rebuttal_2by10', ifsave =True)
    dismatrix_list.append(dismatrix)
    mean_loss = []
    for i in range(200):
        mean_loss.append(dismatrix[i,i])
    print("Experiment No.{} mean loss".format(idx), sum(mean_loss)/200)
        
```
